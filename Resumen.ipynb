```python
#Conjuntos -> SIEMPRE ESTAN ORDENADOS -> SI MODIFICAS UNO SE MODIFICAN LOS DOS A MENOS QUE USES COPY
conjunto = set() #Creas un conjunto vacio
conjunto = {1,2,3} #Creas un conjunto
conjunto.add(4) #Añade 4 al conjunto -> NO ES NECESARIO EL INDICE YA QUE LO COLOCA DE MANERA ORDENADA
conjunto.discard(1) #ELIMINA EL 1 del conjunto
c2 = conjunto.copy() # ESTO PARA CREAR UN CONJUNTO 2 IGUAL AL CONJUNTO 1 Y MODIFICAR EL 2 SIN QUE SE MODIFIQUE EL 1
conjunto.clear() #Vacia el conjunto
conjunto.isdisjoint(c2) #Devuelve true si NO tienen un elemento en comun
conjunto.issubset(c2) #Devuelve true si c1 es subconjunto de c2 -> todo lo que tiene c1 tambien lo tiene c2
conjunto.issuperset(c2) #Devuelve true si c1 contiene a c2 y mas cosas 
conjunto.union(c2) #UNION DE CONJUNTOS -> pone todo lo que hay en los dos conjuntos pero sin contar los repetidos
conjunto.update(c2) #ES LO MISMO DE UNION DE CONJUNTOS PERO ESTE LO MODIFICA EN C1
conjunto.difference(c2) #Pone todos los elementos del conjunto 1 que NO esten en el conjunto 2
conjunto.difference_update(c2) #Lo de arriba pero lo modifica en c1
conjunto.intersection(c2) #Pone solo los elementos que esten en c1 y c2 NO TODOS
conjunto.intersection_update(c2)
conjunto.symmetric_difference(c2)#Diferencia simetrica -> es como una interseccion inversa -> pone lo que esta en 1 que no esta en el 2 y lo que esta en el 2 que no esta en el 1
conjunto.symmetric_difference_update()
4 in conjunto #NOS DEVUELVE TRUE SI 4 ESTA EN EL CONJUNTO
```
